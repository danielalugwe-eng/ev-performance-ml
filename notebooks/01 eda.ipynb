```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\user\electric_car_ml\data\raw\electric_vehicles_spec_2025.csv.csv')
df.head()
```

```python
string_columns = list(df.dtypes[df.dtypes == 'object'].index)    
for col in string_columns:
    df[col] = df[col].str.lower().str.replace(' ', '_') 

df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# Separate numeric and categorical columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

print("Numeric columns:", numeric_cols)
print("Categorical columns:", categorical_cols)
```

```python
from sklearn.model_selection import train_test_split

# Stage 1: Split full dataset
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=42)

# Stage 2: Split training into train and validation
df_train, df_val = train_test_split(df_train_full, test_size=0.25, random_state=42)  # 0.25 x 0.8 = 0.2

# Separate target and features
target = 'range_km'

X_train = df_train.drop(columns=[target])
y_train = df_train[target]

X_val = df_val.drop(columns=[target])
y_val = df_val[target]
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error

# Convert features to dicts for DictVectorizer
dv = DictVectorizer(sparse=False)
X_train_dict = X_train.to_dict(orient='records')
X_val_dict = X_val.to_dict(orient='records')

# Fit and transform the training data, transform the validation data
X_train_encoded = dv.fit_transform(X_train_dict)
X_val_encoded = dv.transform(X_val_dict)

# Train Linear Regression
model = LinearRegression()
model.fit(X_train_encoded, y_train)

# Predict on validation set
y_pred_val = model.predict(X_val_encoded)

# Evaluate using RMSE
rmse_val = root_mean_squared_error(y_val, y_pred_val)
print(f"Linear Regression RMSE on validation set: {rmse_val:.2f}")
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import root_mean_squared_error

alphas = [0.01, 0.1, 1, 10, 100]

ridge_results = []
lasso_results = []

# Ridge
for a in alphas:
    model = Ridge(alpha=a)
    model.fit(X_train_encoded, y_train)
    pred = model.predict(X_val_encoded)
    rmse = root_mean_squared_error(y_val, pred)
    ridge_results.append((a, rmse))

# Lasso
for a in alphas:
    model = Lasso(alpha=a, max_iter=10000)
    model.fit(X_train_encoded, y_train)
    pred = model.predict(X_val_encoded)
    rmse = root_mean_squared_error(y_val, pred)
    lasso_results.append((a, rmse))

print("Ridge:", ridge_results)
print("Lasso:", lasso_results)
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import root_mean_squared_error

alphas = [0.001, 0.01, 0.1, 1, 10]
l1_ratios = [0.1, 0.3, 0.5, 0.7, 0.9]

results_elastic = []

for a in alphas:
    for l1 in l1_ratios:
        model = ElasticNet(alpha=a, l1_ratio=l1, max_iter=20000)
        model.fit(X_train_encoded, y_train)

        pred = model.predict(X_val_encoded)
        rmse = root_mean_squared_error(y_val, pred)

        results_elastic.append((a, l1, rmse))

# Sort by RMSE (best first)
results_elastic = sorted(results_elastic, key=lambda x: x[2])

print("Top ElasticNet results (alpha, l1_ratio, RMSE):")
for r in results_elastic[:10]:
    print(r)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import root_mean_squared_error

rf_results = []

n_estimators_list = [200, 400, 600, 800]
max_depth_list = [None, 10, 20, 30]
min_samples_split_list = [2, 5, 10]

for n in n_estimators_list:
    for depth in max_depth_list:
        for split in min_samples_split_list:
            model = RandomForestRegressor(
                n_estimators=n,
                max_depth=depth,
                min_samples_split=split,
                random_state=42,
                n_jobs=-1
            )
            model.fit(X_train_encoded, y_train)
            pred = model.predict(X_val_encoded)
            rmse = root_mean_squared_error(y_val, pred)
            rf_results.append((n, depth, split, rmse))

sorted_rf = sorted(rf_results, key=lambda x: x[-1])[:10]
print("\nTop Random Forest results (n, depth, split, RMSE):")
for r in sorted_rf:
    print(r)
```

```python
from xgboost import XGBRegressor

xgb_results = []

learning_rates = [0.01, 0.05, 0.1]
max_depths = [4, 6, 8, 10]
n_estimators_list = [500, 1000, 1500]
subsamples = [0.7, 0.9, 1.0]

for lr in learning_rates:
    for depth in max_depths:
        for est in n_estimators_list:
            for sub in subsamples:
                model = XGBRegressor(
                    learning_rate=lr,
                    max_depth=depth,
                    n_estimators=est,
                    subsample=sub,
                    colsample_bytree=1.0,
                    objective='reg:squarederror',
                    random_state=42,
                    n_jobs=-1,
                    tree_method='hist'
                )
                model.fit(X_train_encoded, y_train, eval_set=[(X_val_encoded, y_val)], verbose=False)
                pred = model.predict(X_val_encoded)
                rmse = root_mean_squared_error(y_val, pred)
                xgb_results.append((lr, depth, est, sub, rmse))

sorted_xgb = sorted(xgb_results, key=lambda x: x[-1])[:10]
print("\nTop XGBoost results (lr, depth, estimators, subsample, RMSE):")
for r in sorted_xgb:
    print(r)
```

```python
from catboost import CatBoostRegressor

# Identify categorical columns from current X_train
cat_features = X_train.select_dtypes(include='object').columns.tolist()

# Initialize CatBoost model
cat_model = CatBoostRegressor(
    iterations=5000,
    learning_rate=0.05
,
    depth=8,
    loss_function='RMSE',
    random_seed=42,
    verbose=100,
    early_stopping_rounds=200
)

# Fit model on training data, validate on validation set
cat_model.fit(
    X_train, y_train,
    eval_set=(X_val, y_val),
    cat_features=cat_features,
    use_best_model=True
)

# Predict on validation set
y_pred_val = cat_model.predict(X_val)

# Evaluate RMSE using sklearn's root_mean_squared_error
from sklearn.metrics import mean_squared_error
import numpy as np

rmse_val = np.sqrt(mean_squared_error(y_val, y_pred_val))
print(f"CatBoost RMSE on validation set: {rmse_val:.2f}")
```
